```python
import requests
import bs4
from bs4 import BeautifulSoup
```

```python
# Enter the URL that you want to scrape the data from
url = "https://en.wikipedia.org/wiki/List_of_countries_and_dependencies_by_population"
res = requests.get(url)
soup = BeautifulSoup(res.content, "html.parser")
```

```python
# Prints out the HTML Page
# print(soup)
```

```python
table = soup.find("table") # Find table in the webpage

for row in table.find_all("tr")[1:]: # Finds all the table rows
    col = row.find_all("td") # find all data 
    print(col[0].text) # Print col[0] data
```
